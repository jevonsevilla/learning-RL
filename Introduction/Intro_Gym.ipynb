````python
env = gym.make('Pong-v0')
env.reset()

for _ in range(1000):
    env.render()
    env.step(env.action_space.sample())
    
env.close()
````

# Observations
the iterative observations returned by the environment. Obtained through the `.step()` module

[Observation (object), reward (float), done (boolean), info (dict)]

Observation - environement specific object representation of state

Reward - amound of reward achieved by the previous action

Done - indicates if it's time to reset the environment again to a new episode

Info - diagnostic information for debugging. (for example, it might contain the raw probabilities behind the environmentâ€™s last state change)

````python
env = gym.make('CartPole-v0')
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
````
